```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("sym.csv")
```

```python
data.head()
```

```python
plt.figure(figsize = (4, 3),dpi = 600)

x = data["arc_length"]
y1 = data["t100"]
y2 = data["t50"]
y3 = data["q50"]
y4 = data["q100"]

plt.plot(x, y1, label = "t100", lw = 1.)
plt.plot(x, y2, label = "t100", lw = 1.)
plt.plot(x, y3, label = "t100", lw = 1.)
plt.plot(x, y4, label = "t100", lw = 1.)

plt.xlabel('x', fontsize=10.)
plt.ylabel('T', fontsize=10.)

plt.title("    ", fontsize=8.)
plt.legend(fontsize=6.)
plt.savefig("sym.jpg", dpi = 300)
plt.show()
```
