```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize = (8, 4.5),dpi = 450)

x = np.array([1, 2, 4])
y = np.array([1., 34.0/21.8, 34.0/15.84])

plt.plot(x, y, label = "6000 Mb", lw = 1.)

plt.xlabel("Число узлов", fontsize=10.)
plt.ylabel("Ускорение",   fontsize=10.)

plt.title("Параллельные вычисления", fontsize=8.)
plt.legend(fontsize=6.)
plt.savefig("5.jpg", dpi = 450)
plt.show()
```
